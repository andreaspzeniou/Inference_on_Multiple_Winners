```python
import numpy as np
import pandas as pd
from itertools import product
import matplotlib.pyplot as plt
import scipy as scp
import multiprocessing
```

```python
df = pd.read_csv("Final_Simulation_Study.csv")
```

```python
df.head()
```

```python
df_p_cov = pd.DataFrame()
df_ts_cov = pd.DataFrame()
# df_LS1_cov = pd.DataFrame()
df_LS2_cov = pd.DataFrame()
df_zoom_cov = pd.DataFrame()

df_p_len = pd.DataFrame()
df_ts_len = pd.DataFrame()
# df_LS1_len = pd.DataFrame()
df_LS2_len = pd.DataFrame()
df_zoom_len = pd.DataFrame()

for i in range(5):
    
    df_p_cov["cov_prob" + str(10 ** i)] = np.array([df["p_cov_prob"].iloc[j].replace("[", "").replace("]", "").split(" ")[i] for j in range(len(df["p_cov_prob"]))]).astype("float") / 10000
    df_ts_cov["cov_prob" + str(10 ** i)] = np.array([df["ts_cov_prob"].iloc[j].replace("[", "").replace("]", "").split(" ")[i] for j in range(len(df["ts_cov_prob"]))]).astype("float") / 10000
    # df_LS1_cov["cov_prob" + str(10 ** i)] = np.array([df["LS1_cov_prob"].iloc[j].replace("[", "").replace("]", "").split(" ")[i] for j in range(len(df["LS1_cov_prob"]))]).astype("float") / 10000
    df_LS2_cov["cov_prob" + str(10 ** i)] = np.array([df["LS2_cov_prob"].iloc[j].replace("[", "").replace("]", "").split(" ")[i] for j in range(len(df["LS2_cov_prob"]))]).astype("float") / 10000
    df_zoom_cov["cov_prob" + str(10 ** i)] = np.array([df["zoom_cov_prob"].iloc[j].replace("[", "").replace("]", "").split(" ")[i] for j in range(len(df["zoom_cov_prob"]))]).astype("float") / 10000
    
    df_p_len["length" + str(10 ** i)] = np.array([(" ".join(df["p_length"].iloc[j].replace("[", "").replace("]", "").split())).split(" ")[i] for j in range(len(df["p_length"]))]).astype("float") / 10000
    df_ts_len["length" + str(10 ** i)] = np.array([(" ".join(df["ts_length"].iloc[j].replace("[", "").replace("]", "").split())).split(" ")[i] for j in range(len(df["ts_length"]))]).astype("float") / 10000
    # df_LS1_len["length" + str(10 ** i)] = np.array([(" ".join(df["LS1_length"].iloc[j].replace("[", "").replace("]", "").split())).split(" ")[i] for j in range(len(df["LS1_length"]))]).astype("float") / 10000
    df_LS2_len["length" + str(10 ** i)] = np.array([(" ".join(df["LS2_length"].iloc[j].replace("[", "").replace("]", "").split())).split(" ")[i] for j in range(len(df["LS2_length"]))]).astype("float") / 10000
    df_zoom_len["length" + str(10 ** i)] = np.array([(" ".join(df["zoom_length"].iloc[j].replace("[", "").replace("]", "").split())).split(" ")[i] for j in range(len(df["zoom_length"]))]).astype("float") / 10000
    
df_p_cov["Design"] = df["Design"]
df_ts_cov["Design"] = df["Design"]
# df_LS1_cov["Design"] = df["Design"]
df_LS2_cov["Design"] = df["Design"]
df_zoom_cov["Design"] = df["Design"]

df_p_len["Design"] = df["Design"]
df_ts_len["Design"] = df["Design"]
# df_LS1_len["Design"] = df["Design"]
df_LS2_len["Design"] = df["Design"]
df_zoom_len["Design"] = df["Design"]

df_p_cov["Corr"] = df["Corr"]
df_ts_cov["Corr"] = df["Corr"]
# df_LS1_cov["Corr"] = df["Corr"]
df_LS2_cov["Corr"] = df["Corr"]
df_zoom_cov["Corr"] = df["Corr"]

df_p_len["Corr"] = df["Corr"]
df_ts_len["Corr"] = df["Corr"]
# df_LS1_len["Corr"] = df["Corr"]
df_LS2_len["Corr"] = df["Corr"]
df_zoom_len["Corr"] = df["Corr"]

df_p_cov["CI"] = "P"
df_ts_cov["CI"] = "TS"
# df_LS1_cov["CI"] = "LS1"
df_LS2_cov["CI"] = "LS2"
df_zoom_cov["CI"] = "zoom"

df_p_len["CI"] = "P"
df_ts_len["CI"] = "TS"
# df_LS1_len["CI"] = "LS1"
df_LS2_len["CI"] = "LS2"
df_zoom_len["CI"] = "zoom"
```

```python
df_cov_p = pd.concat([df_p_cov, df_ts_cov, df_LS1_cov, df_LS2_cov, df_zoom_cov])
df_cov_p["idx"] = df_cov_p.index
df_cov_p = df_cov_p.sort_values(["idx", "CI"])

df_len = pd.concat([df_p_len, df_ts_len, df_LS1_len, df_LS2_len, df_zoom_len])
df_len["idx"] = df_len.index
df_len = df_len.sort_values(["idx", "CI"])
```

```python
def get_mu_quantile(val, l, r, mu, s):
    return(scp.stats.truncnorm.cdf(val, ((l - mu) / s), ((r - mu) / s), loc = mu, scale = s))
```

```python
def cond_inference(Y, X, Sigma, th_hat, c, alpha, mu_grid = []):
    
    p = len(Y)
    
    Sigma_Y = Sigma[ : p, : p]
    Sigma_XY = Sigma[ : p, p :]
    
    Z = []
    l = []
    
    for i in range(N):
        Z.append(X[i] - Sigma_XY[i, th_hat] * Y[th_hat] / Sigma_Y[th_hat, th_hat])
        l.append(Sigma_XY[i, th_hat] / Sigma_Y[th_hat, th_hat])
        
    N_p = []
    N_m = []
    
    for i in range(N):
        
        if i != th_hat:
            if l[i] > 0:
                N_p.append((c - Z[i]) / l[i])
            elif l[i] < 0:
                N_m.append((c - Z[i]) / l[i])
                
        elif i == th_hat:
            if l[i] > 0:
                N_m.append((c - Z[i]) / l[i])
            elif l[i] < 0:
                N_p.append((c - Z[i]) / l[i])
    
    try: N_m = max(N_m)
    except: N_m = min(mu_grid) - 10 * Sigma_Y[th_hat, th_hat]
        
    try: N_p = min(N_p)
    except: N_p = max(mu_grid) + 10 * Sigma_Y[th_hat, th_hat]
    
    values = []
            
    for mu in mu_grid:
        values.append(get_mu_quantile(Y[th_hat], N_m, N_p, mu, Sigma_Y[th_hat, th_hat] ** 0.5))
        
    values_l = np.array(values) - (1 - alpha / 2)
    values_r = np.array(values) - alpha / 2
    
    l_i = np.argmin(abs(values_l))
    r_i = np.argmin(abs(values_r))
    
    l = mu_grid[l_i]
    r = mu_grid[r_i]
    
    return([l,r])
    
```

```python
def hybrid_inference(Y, X, Sigma, th_hat, c, alpha, beta, mu_grid = []):
    
    p = len(Y)
    
    Sigma_Y = Sigma[ : p, : p]
    Sigma_XY = Sigma[ : p, p :]
    
    Z = []
    l = []
    
    for i in range(N):
        Z.append(X[i] - Sigma_XY[i, th_hat] * Y[th_hat] / Sigma_Y[th_hat, th_hat])
        l.append(Sigma_XY[i, th_hat] / Sigma_Y[th_hat, th_hat])
        
    N_p = []
    N_m = []
    
    for i in range(N):
        
        if i != th_hat:
            if l[i] > 0:
                N_p.append((c - Z[i]) / l[i])
            elif l[i] < 0:
                N_m.append((c - Z[i]) / l[i])
                
        elif i == th_hat:
            if l[i] > 0:
                N_m.append((c - Z[i]) / l[i])
            elif l[i] < 0:
                N_p.append((c - Z[i]) / l[i])
    
    try: N_m = max(N_m)
    except: N_m = min(mu_grid) - 10 * Sigma_Y[th_hat, th_hat]
        
    try: N_p = min(N_p)
    except: N_p = max(mu_grid) + 10 * Sigma_Y[th_hat, th_hat]
        
    c_q = c_quantile(1 - beta, Sigma_Y, 1000, np.random.default_rng())
    
    values = []
            
    for mu in mu_grid:
        N_m_mu = max([N_m, mu - Sigma_Y[th_hat, th_hat] ** 0.5 * c_q])
        N_p_mu = min([N_p, mu + Sigma_Y[th_hat, th_hat] ** 0.5 * c_q])
        if N_p_mu < N_m_mu:
            if mu - Sigma_Y[th_hat, th_hat] ** 0.5 * c_q > N_p_mu:
                values.append(0)
            else:
                values.append(1)
        else:
            values.append(get_mu_quantile(Y[th_hat], N_m_mu, N_p_mu, mu, Sigma_Y[th_hat, th_hat] ** 0.5))
        
    gamma = (alpha - beta) / (1 - beta)
        
    values_l = np.array(values) - (1 - gamma / 2)
    values_r = np.array(values) - gamma / 2
    
    l_i = np.argmin(abs(values_l))
    r_i = np.argmin(abs(values_r))
    
    l = mu_grid[l_i]
    r = mu_grid[r_i]
    
    return([l,r])
    
```

```python
def c_quantile(a, Sigma_Y, n_iter, rand):

    p = len(Sigma_Y)
    
    draws = rand.multivariate_normal(np.zeros(p), Sigma_Y, n_iter)
    diags = np.tile((np.diagonal(Sigma_Y) ** (1/2)), (n_iter, 1))
    draws2 = abs(draws / diags)
    
    draws = np.where((diags != 0), draws2, abs(draws))
    
    lst = np.amax(draws, axis=1)
    return np.quantile(np.array(lst), a)
```

```python
def c_quantile_local(a, Sigma_Y, J, n_iter, rand):

    Sigma_Y = Sigma_Y[J,J]
    p = len(Sigma_Y)
    
    draws = abs(rand.multivariate_normal(np.zeros(p), Sigma_Y, n_iter))
    diags = np.tile((np.diagonal(Sigma_Y) ** (1/2)), (n_iter, 1))
    draws2 = abs(draws / diags)
    
    draws = np.where((diags != 0), draws2, abs(draws))
    
    lst = np.amax(draws, axis=1)
    return np.quantile(np.array(lst), a)
```

```python
def d_quantile(b, k, Sigma_X, n_iter, rand):

    N = len(Sigma_X)
    
    draws = rand.multivariate_normal(np.zeros(N), Sigma_X, n_iter)
    lst = np.amax(draws, axis=1) - np.amax((-1 * draws), axis=1)

    return np.quantile(np.array(lst), b)
```

```python
def make_I_list(rank_list, X_temp, delta_L, delta_U):
    
    rank = np.max(np.array(rank_list))
    
    X_temp_mat = np.tile(X_temp, (len(X_temp), 1))
    A = X_temp_mat + np.transpose(delta_L)
    
    np.fill_diagonal(A, np.nan)
    A = A[~np.isnan(A)].reshape(A.shape[0], A.shape[1] - 1)

    row_ranks = np.partition(A, -rank, axis=1)[:, -rank]
    
    indices = np.where(X_temp >= row_ranks, 1, 0)
        
    return indices
```

```python
def c_2_quantile(ab, rank_list, Sigma, delta_L, delta_U, n_iter, rand):

    lst = []
    
    k = len(delta_L)
    p = int(len(Sigma) / (k + 1))
    
    Sigma_Y = Sigma[: p, : p]

    for a in range(n_iter):
        draw = rand.multivariate_normal(np.zeros(p * (k + 1)), Sigma)
        
        I_list = []
        
        for i in range(k):
                                        
            X_temp = draw.copy()
            X_temp = X_temp[((i + 1) * p) : ((i + 2) * p)]
            rank = np.max(np.array(rank_list))
            
            I_list.append(make_I_list([j+1 for j in range(rank_list[i])], X_temp, delta_L[i], delta_U[i]))

        s = []
        
        Y_temp = draw.copy()
        Y_temp = Y_temp[:p]
        
        all_indices = []
                                        
        for i in range(k):
                                        
            I_l = I_list[i]
            I_indices = np.where(I_l == 1)[0]
            all_indices = list(set(all_indices) | set(I_indices))
                
        for j in all_indices:
            if Sigma_Y[j,j] > 0:
                s.append(abs(Y_temp[j] / (Sigma_Y[j,j] ** 0.5)))
            else:
                s.append(0)
        
        lst.append(np.max(np.array(s)))

    return np.quantile(np.array(lst), ab)
```

```python
def c_2_quantile_asymmetric(ab, rank_list, Sigma, delta_L, delta_U, n_iter, rand):
    
    ab_p = 1 - ab
    ab_p = ab_p / 2
    ab_p = 1 - ab_p

    lst_l = []
    lst_r = []
    lst = []
    
    k = len(delta_L)
    p = int(len(Sigma) / (k + 1))
    
    Sigma_Y = Sigma[: p, : p]

    for a in range(n_iter):
        draw = rand.multivariate_normal(np.zeros(p * (k + 1)), Sigma)
        
        I_list = []
        
        for i in range(k):
                                        
            X_temp = draw.copy()
            X_temp = X_temp[((i + 1) * p) : ((i + 2) * p)]
            rank = np.max(np.array(rank_list))
            
            I_list.append(make_I_list([j+1 for j in range(rank_list[i])], X_temp, delta_L[i], delta_U[i]))

        s = []
        
        Y_temp = draw.copy()
        Y_temp = Y_temp[:p]
        
        all_indices = []
                                        
        for i in range(k):
                                        
            I_l = I_list[i]
            I_indices = np.where(I_l == 1)[0]
            all_indices = list(set(all_indices) | set(I_indices))
                
        for j in all_indices:
            if Sigma_Y[j,j] > 0: s.append(Y_temp[j] / (Sigma_Y[j,j] ** 0.5))
            else: s.append(0)
        
        lst_r.append(np.max(np.array(s)))
        lst_l.append(np.min(np.array(s)))
        lst.append(np.max(abs(np.array(s))))

    return([np.quantile(-1 * np.array(lst_l), ab_p), np.quantile(np.array(lst_r), ab_p)])
```

```python
# Andrews et al. projection CI

def andrews_CI_proj(rank_list, x_vec_list, y_vec, a, Sigma_Y, n_iter, quantile_dict, rand):

    th_hat_lst = []
    
    for i in range(len(x_vec_list)):
        x_vec = x_vec_list[i]
        for j in range(rank_list[i]):
            th_hat_lst.append(np.argsort(x_vec)[len(x_vec)-(j+1)])
            
    c_a = c_quantile(a, Sigma_Y, n_iter, rand = rand)
        
    CI_list = []
        
    for i in range(len(th_hat_lst)):  

        CI_lower_bound = y_vec[th_hat_lst[i]] - c_a * Sigma_Y[th_hat_lst[i], th_hat_lst[i]] ** 0.5
        CI_upper_bound = y_vec[th_hat_lst[i]] + c_a * Sigma_Y[th_hat_lst[i], th_hat_lst[i]] ** 0.5
        
        CI_list.append([CI_lower_bound, CI_upper_bound])

    return CI_list

```

```python
def two_step_CI(rank_list, x_vec_lst, y_vec, a, b, Sigma, n_iter, quantile_dict, rand):

    N = len(y_vec)
    k = len(x_vec_lst)
    
    Sigma_X = Sigma[N :, N :]
    Sigma_Y = Sigma[: N, : N]

    th_hat_lst = [] 
    
    for i in range(len(x_vec_lst)):
        x_vec = x_vec_lst[i]
        for j in range(rank_list[i]):
            th_hat_lst.append(np.argsort(x_vec)[len(x_vec)-(j+1)])

    L = [np.zeros((N, N)) for i in range(k)]
    U = [np.zeros((N, N)) for i in range(k)]
    d = d_quantile(1 - b, k, Sigma_X, n_iter, rand = rand)
    
    for i in range(len(x_vec_lst)):
        
        x_vec = x_vec_lst[i]
        l = np.tile(x_vec, (len(x_vec), 1))
        r = np.transpose(l)

        L[i] = r - l - d
        U[i] = r - l + d

    bd = c_2_quantile(1 - a + b, rank_list, Sigma, L, U, int(n_iter), rand = rand)
    
    CI_list = []
        
    for i in range(len(th_hat_lst)):  

        CI_lower_bound = y_vec[th_hat_lst[i]] - bd * (Sigma_Y[th_hat_lst[i], th_hat_lst[i]] ** 0.5)
        CI_upper_bound = y_vec[th_hat_lst[i]] + bd * (Sigma_Y[th_hat_lst[i], th_hat_lst[i]] ** 0.5)
        
        CI_list.append([CI_lower_bound, CI_upper_bound])

    return CI_list
```

```python
def two_step_CI_asymmetric(rank_list, x_vec_lst, y_vec, a, b, Sigma, n_iter, quantile_dict, rand):

    N = len(y_vec)
    k = len(x_vec_lst)
    
    Sigma_X = Sigma[N :, N :]
    Sigma_Y = Sigma[: N, : N]

    th_hat_lst = [] 
    
    for i in range(len(x_vec_lst)):
        x_vec = x_vec_lst[i]
        for j in range(rank_list[i]):
            th_hat_lst.append(np.argsort(x_vec)[len(x_vec)-(j+1)])

    L = [np.zeros((N, N)) for i in range(k)]
    U = [np.zeros((N, N)) for i in range(k)]
    d = d_quantile(1 - b / 2, k, Sigma_X, int(n_iter), rand = rand)
    
    for i in range(len(x_vec_lst)):
        
        x_vec = x_vec_lst[i]
        l = np.tile(x_vec, (len(x_vec), 1))
        r = np.transpose(l)

        L[i] = r - l - d
        U[i] = r - l + d

    bd = c_2_quantile_asymmetric(1 - a + b, rank_list, Sigma, L, U, int(n_iter), rand = rand)
    bd_l = bd[0]
    bd_r = bd[1]
    
    CI_list = []
        
    for i in range(len(th_hat_lst)):  

        CI_lower_bound = y_vec[th_hat_lst[i]] - bd_r * (Sigma_Y[th_hat_lst[i], th_hat_lst[i]] ** 0.5)
        CI_upper_bound = y_vec[th_hat_lst[i]] + bd_l * (Sigma_Y[th_hat_lst[i], th_hat_lst[i]] ** 0.5)
        
        CI_list.append([CI_lower_bound, CI_upper_bound])

    return CI_list
```

```python
# Second locally-simultaneous CI

def LS2_CI(rank_list, x_vec_lst, y_vec, a, b, Sigma, n_iter, quantile_dict, rand):

    th_hat_lst = [] 
    
    N = len(y_vec)
    k = len(x_vec_lst)
    
    Sigma_X = Sigma[N :, N :]
    Sigma_Y = Sigma[: N, : N]
    
    for i in range(len(rank_list)):
        th_hat_lst.append(np.argsort(x_vec_lst[i])[len(x_vec_lst[i])-rank_list[i]])
        
    # want to construct J_hat
    
    J_hat = np.array([])
    
    d = d_quantile(1 - b, k, Sigma_X, n_iter, rand = rand)
    
    for i in range(len(rank_list)):

        lb = x_vec_lst[i][th_hat_lst[i]] - d
        J_hat = np.union1d(np.where(x_vec_lst[i] >= lb), J_hat)
        
    J_hat = J_hat.astype(int)
    
    Sigma_YR = Sigma_Y[J_hat,:]
    Sigma_YR = Sigma_YR[:,J_hat]
    
    c_a = c_quantile(1 - (a - b), Sigma_YR, int(n_iter), rand)
    c_a2 = c_quantile(1 - a, Sigma_Y, int(n_iter), rand)
    
    c_a = min([c_a, c_a2])
    
    th_hat_lst = [] 
    
    for i in range(len(x_vec_lst)):
        x_vec = x_vec_lst[i]
        for j in range(rank_list[i]):
            th_hat_lst.append(np.argsort(x_vec)[len(x_vec)-(j+1)])
        
    CI_list = []
        
    for i in range(len(th_hat_lst)):  

        CI_lower_bound = y_vec[th_hat_lst[i]] - c_a * (Sigma_Y[th_hat_lst[i], th_hat_lst[i]] ** 0.5)
        CI_upper_bound = y_vec[th_hat_lst[i]] + c_a * (Sigma_Y[th_hat_lst[i], th_hat_lst[i]] ** 0.5)
        
        CI_list.append([CI_lower_bound, CI_upper_bound])

    return CI_list
```

```python
def r_l(y, rank_list, x_vec_lst, a, var_vec):
    
    x_vec = x_vec_lst[0]
    
    rank = np.max(np.array(rank_list))
    th_hat_lst = []
    for j in range(rank):
        th_hat_lst.append(np.argsort(x_vec)[len(x_vec)-(j+1)])
    
    m = len(y)
    
    th_hat = np.argmax(y)
    delta_hat = []
    for j in range(len(y)):
        delta_hat.append(max([0, y[th_hat_lst[-1]] - y[j]]))
        
    delta_hat = np.array(delta_hat)
    delta_hat = np.flip(np.sort(delta_hat))
    
    alpha = a
    
    for j in range(len(y)):
        r_l_j = (-1 * np.max(var_vec ** 0.5)) * scp.stats.norm.ppf(alpha / (2 * (m - j)))
        print(r_l_j)
        if delta_hat[j] <= 4 * r_l_j:
            return r_l_j
        else:
            alpha = alpha - 2 * (scp.stats.norm.cdf(-1 * (delta_hat[j] - r_l_j) / 3) * (1 / np.max(var_vec ** 0.5)))
    
```

```python
def zrnic_method2(rank_list, x_vec_lst, y_vec, a, Sigma, n_iter, quantile_dict, rand):

    th_hat_lst = [] 
    
    N = len(y_vec)
    k = len(x_vec_lst)
    
    x_vec = x_vec_lst[k-1]
    
    Sigma_X = Sigma[N :, N :]
    Sigma_Y = Sigma[: N, : N]
    
    var_vec = np.diag(Sigma_Y)
    
    rank = np.max(np.array(rank_list))
    for j in range(rank):
        th_hat_lst.append(np.argsort(x_vec)[len(x_vec)-(j+1)])
        
    r_l_hat = r_l(y_vec, rank_list, x_vec_lst, a, var_vec)
    
    return([[y_vec[th_hat] - r_l_hat, y_vec[th_hat] + r_l_hat] for th_hat in th_hat_lst])
    
```

```python
def individual_run_one(param_list, q_dict = {}):

    if len(param_list) == 6:

        mu_vec_x, mu_vec_y, Sigma, n_test, s_level, seed = param_list[0], param_list[1], param_list[2], param_list[3], param_list[4], param_list[5]
        
    k = len(mu_vec_y)
    N = int(len(mu_vec_x) / k)

    mu_vec_x_list = []

    for i in range(N):
        mu_vec_x_list.append(mu_vec_x[i * k : (i + 1) * k])
        
    rng = np.random.default_rng(seed)
    mu_vec = np.concatenate((mu_vec_y, mu_vec_x))
    
    draw = rng.multivariate_normal(np.zeros(len(mu_vec)), Sigma)
    
    std_cov_p = 0
    p_cov_p = 0
    shaikh_cov_p = 0
    LS2_cov_p = 0
    
    std_l = 0
    p_l = 0
    shaikh_l = 0
    LS2_l = 0
    
    a_vec = draw + mu_vec

    y_vec = a_vec[:k]
    x_vec = a_vec[k:]
    
    x_list = []
    Y = y_vec
    
    for i in range(N):
        x_list.append(x_vec[i * k : (i + 1) * k])

    p_CI = andrews_CI_proj([13], x_list, Y, 0.95, Sigma[:len(Y), :len(Y)], n_test, {}, rng)
    sh_CI = two_step_CI([13], x_list, Y, 0.05, 0.005, Sigma, n_test, {}, rng)
    LS2_CI_out = LS2_CI([13], x_list, Y, 0.05, 0.005, Sigma, n_test, {}, rng)

    rank_list = [13]
    
    p_add = 1
    shaikh_add = 1
    std_add = 1
    LS2_add = 1

    p_vol = 1
    shaikh_vol = 1
    std_vol = 1
    LS2_vol = 1
    
    Sigma_Y = Sigma[:len(Y), :len(Y)]
    
    th_hat_list = []
    th_hat_list_tested = []
    
    n_tested = 0

    for i in range(len(rank_list)):
        for j in range(rank_list[i]):
            th_hat_list.append(np.argsort(x_list[i])[len(x_list[i])-(j+1)])
            
            if np.sqrt(Sigma_Y[th_hat_list[-1], th_hat_list[-1]]) > 0: 
                n_tested += 1
                th_hat_list_tested.append(np.argsort(x_list[i])[len(x_list[i])-(j+1)])
                
    if n_tested > 0:

        Sigma_Y = Sigma[:len(Y), :len(Y)]
        Sigma_Y2 = Sigma_Y[th_hat_list_tested,:]
        Sigma_Y2 = Sigma_Y2[:,th_hat_list_tested]

        std_CI = andrews_CI_proj([n_tested], [Y[th_hat_list_tested]], Y[th_hat_list_tested], 0.95, Sigma_Y2, n_test, {}, rng)

        for i in range(len(th_hat_list)):

            th_hat = th_hat_list[i]

            if np.sqrt(Sigma_Y[th_hat, th_hat]) > 0:

                if (p_CI[i][0] <= mu_vec_y[th_hat] <= p_CI[i][1]): p_add *= 1
                else: p_add *= 0

                if (sh_CI[i][0] <= mu_vec_y[th_hat] <= sh_CI[i][1]): shaikh_add *= 1
                else: shaikh_add *= 0
                    
                if (LS2_CI_out[i][0] <= mu_vec_y[th_hat] <= LS2_CI_out[i][1]): LS2_add *= 1
                else: LS2_add *= 0

                p_vol *= p_CI[i][1] - p_CI[i][0]
                shaikh_vol *= sh_CI[i][1] - sh_CI[i][0]
                LS2_vol *= LS2_CI_out[i][1] - LS2_CI_out[i][0]

        for i in range(len(th_hat_list_tested)):

            th_hat = th_hat_list_tested[i]

            if np.sqrt(Sigma_Y[th_hat, th_hat]) > 0:

                if (std_CI[i][0] <= mu_vec_y[th_hat] <= std_CI[i][1]): p_add *= 1
                else: std_add *= 0

                std_vol *= std_CI[i][1] - std_CI[i][0]

        p_cov_p += p_add
        std_cov_p += std_add
        shaikh_cov_p += shaikh_add
        LS2_cov_p += LS2_add

        p_l += p_vol
        std_l += std_vol
        shaikh_l += shaikh_vol
        LS2_l += LS2_vol

    return (p_cov_p,
            std_cov_p,
            shaikh_cov_p,
            LS2_cov_p,
            p_l,
            std_l,
            shaikh_l,
            LS2_l,
            n_tested)
     
```
